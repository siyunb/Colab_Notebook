```python
%matplotlib inline
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader 
from torch import optim
import numpy as np
import pandas as pd
from collections import OrderedDict 
import sys
from IPython import display
from matplotlib import pyplot as plt
sys.path.append('..')

torch.set_default_tensor_type(torch.FloatTensor)
```

```python
train_data = pd.read_csv('./Dataset/HousePrices/train.csv')
test_data = pd.read_csv('./Dataset/HousePrices/test.csv')
```

```python
all_data = pd.concat([train_data.iloc[:,1:-1], test_data.iloc[:, 1:]],axis = 0)
numeric_column = all_data.dtypes[all_data.dtypes != 'object'].index
```

```python
all_data[numeric_column] = all_data[numeric_column].apply(lambda x:(x - x.mean())/x.std())
```

```python
all_data[numeric_column] = all_data[numeric_column].fillna(0)
```

```python
all_data = pd.get_dummies(all_data, dummy_na=True)
all_data.shape 
```

```python
n_train = len(train_data)
tensor_train_data = torch.tensor(all_data[:n_train].values) 
tensor_test_data = torch.tensor(all_data[n_train:].values)
tensor_price = torch.tensor(train_data['SalePrice'].values).unsqueeze(1)
```

```python
#initialate parameters
def initate():
    num_inputs = 331
    num_hiddens1 = 124
    num_hiddens2 = 56
    num_outputs = 1

    w1 = torch.tensor(np.random.normal(0, 0.01, (num_inputs, num_hiddens1)), requires_grad = True, dtype = torch.float)
    b1 = torch.zeros(1, requires_grad = True, dtype = torch.float)
    w2 = torch.tensor(np.random.normal(0, 0.01, (num_hiddens1, num_hiddens2)),requires_grad = True, dtype = torch.float)
    b2 = torch.zeros(1, requires_grad = True, dtype = torch.float)
    w3 = torch.tensor(np.random.normal(0, 0.01, (num_hiddens2, num_outputs)),requires_grad = True, dtype = torch.float)
    b3 = torch.zeros(1, requires_grad = True, dtype = torch.float)

    params = [w1, b1, w2, b2, w3, b3]
    return params
```

```python
def dropout(X, drop_prob):
    X = X.float()
    assert 0<=drop_prob<=1
    keep_prob = 1- drop_prob
    if keep_prob == 0:
        return torch.zeros_like(X)
    mask = (torch.rand(X.shape,dtype = torch.float)<keep_prob).float()
    return mask*X/keep_prob
```

```python
def network(X, params, drop_prob, is_training = True):
    X = X.float()
    w1, b1, w2, b2, w3, b3 = params
    h1 = (torch.matmul(X, w1) + b1).relu()
    if is_training:
        h1 = dropout(h1, drop_prob)
    h2 = (torch.matmul(h1, w2) + b2).relu()
    if is_training:
        h2 = dropout(h2, drop_prob)    
    o = torch.matmul(h2, w3) + b3
    return o
```

```python
def loss_mse(y_hat, y):
    return (y_hat - y).float().pow(2).mean()
```

```python
def log_mse_z(y_hat, y):
    with torch.no_grad():
        clipped_preds = torch.max(y_hat, torch.tensor(1.0, dtype = torch.float))
        mse = torch.sqrt((clipped_preds.log()- y.log()).pow(2).mean())
    return mse.item()
```

```python
def training_loop(num_epoch_c, batch_size_c, lr_c, drop_prob_c, wd_c,
                  params_c, model_fn, loss_fn,
                  train_x_c, train_y_c, test_x_c = None, test_y_c = None):
    dataset = TensorDataset(train_x_c, train_y_c)
    train_iter = DataLoader(dataset, batch_size_c, shuffle = True)
    train_loss, test_loss = [], []
    optimizer = optim.Adam(params = params_c, lr= lr_c, weight_decay= wd_c)
    for epoch in range(num_epoch_c):
        for X, y in train_iter:
            y_hat = model_fn(X, params_c, drop_prob_c)
            loss = loss_fn(y_hat, y)
            
            optimizer.zero_grad()

            loss.backward()
                        
            optimizer.step()
        
        train_loss.append(log_mse_z(model_fn(train_x_c, params_c, drop_prob_c, is_training = False), train_y_c))
        
        if test_x_c is not None:
            test_loss.append(log_mse_z(model_fn(test_x_c, params_c, drop_prob_c, is_training = False), test_y_c))
    return params_c, train_loss, test_loss
```

```python
def k_fold_ego(k_c,
               num_epoch_c, batch_size_c, lr_c, drop_prob_c, wd_c,
               tensor_train_data_c, tensor_price_c,
               model_fn, loss_fn):
    train_loss_sum, valid_loss_sum = 0, 0
    for i in range(k_c):
        params = initate()
        data = get_k_fold_data(k_c, i, tensor_train_data_c, tensor_price_c)
        params, train_loss, valid_loss = training_loop(num_epoch_c, batch_size_c, lr_c, drop_prob_c, wd_c,
                                                       params, model_fn, loss_fn,
                                                       *data)
        train_loss_sum += train_loss[-1]
        valid_loss_sum += valid_loss[-1]
        if i == 0:
            semilogy(range(1, num_epoch_c+1), train_loss, 'epochs', 'rmse',
                    range(1,num_epoch_c+1), valid_loss, ['train','valid'])
        print('fold: %d, train_rmse: %.2f, valid_rmse: %.2f'%(i, train_loss[-1], valid_loss[-1]))
    return train_loss_sum/k_c, valid_loss_sum/k_c   
```

```python
params = initate()
params, _, _i = training_loop(num_epoch_c = 100, batch_size_c = 70, lr_c = 0.2, drop_prob_c = 0.3, wd_c = 0.065,
                  params_c = params, model_fn = network, loss_fn = loss_mse,
                  train_x_c = tensor_train_data, train_y_c = tensor_price, test_x_c = None, test_y_c = None)
```

```python
y_pred = network(tensor_test_data, params, drop_prob = 0.3, is_training = False).detach().squeeze().numpy()
test_data['SalePrice'] = pd.Series(y_pred)
submission = pd.concat([test_data['Id'],test_data['SalePrice']], axis = 1)
submission.to_csv('./Result/submission.csv', index = False)
```

```python
def get_net(num_inputs):
    network = nn.Linear(num_inputs, 1)
    for param in network.parameters():
        nn.init.normal_(param, mean = 0, std = 0.01)
    return network.float()
```

```python
def log_mse(network, X, y):
    with torch.no_grad():
        clipped_preds = torch.max(network(X.float()), torch.tensor(1.0, dtype = torch.float))
        mse = torch.sqrt((clipped_preds.log()- y.log()).pow(2).mean())
    return mse.item()
```

```python
def train(num_epoch, batch_size, model_fn, loss_fn, lr, wd, train_x, train_y, test_x = None, test_y = None):
    dataset = TensorDataset(train_x, train_y)
    train_iter = DataLoader(dataset, shuffle = True, batch_size = batch_size)
    optimizer = optim.Adam(params = model_fn.parameters(), lr= lr, weight_decay= wd)
    train_loss, test_loss = [], []
    for epoch in range(num_epoch):
        for X, y in train_iter:
            loss = loss_fn(model_fn(X.float()), y.float())
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        train_loss.append(log_mse(model_fn, train_x, train_y))
        if test_y is not None:
            test_loss.append(log_mse(model_fn, test_x, test_y))        
    return train_loss, test_loss
```

```python
def get_k_fold_data(k, i, train_x, train_y):
    assert k>1
    fold_size = len(train_y)//k
    train_x_k, train_y_k = None, None 
    for j in range(k):
        idx = slice(j*fold_size, (j+1)*fold_size)
        if j == i:
            valid_x_k, valid_y_k = train_x[idx,:], train_y[idx]
        elif train_x_k is None:
            train_x_k, train_y_k = train_x[idx,:], train_y[idx]
        else:
            train_x_k = torch.cat((train_x_k, train_x[idx,:]), dim = 0)
            train_y_k = torch.cat((train_y_k, train_y[idx]),dim = 0)
    return train_x_k, train_y_k, valid_x_k, valid_y_k
```

```python
#plot function
def set_figsize(figsize = (3.5, 2.5)):
    display.set_matplotlib_formats('svg')
    plt.rcParams['figure.figsize'] = figsize

def semilogy(x_vals, y_vals, x_label, y_label, x2_vals=None, y2_vals=None,
             legend=None, figsize=(3.5, 2.5)):
    set_figsize(figsize)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.semilogy(x_vals, y_vals)
    if x2_vals and y2_vals:
        plt.semilogy(x2_vals, y2_vals, linestyle=':')
        plt.legend(legend)
```

```python
def k_fold(k, train_x, train_y, num_epoch, lr, wd, batch_size):
    train_loss_sum, valid_loss_sum = 0, 0
    for i in range(k):
        model_fn = get_net(train_x.shape[1])
        loss_fn = nn.MSELoss()
        data = get_k_fold_data(k, i, train_x, train_y)
        train_loss, valid_loss = train(num_epoch, batch_size, model_fn, loss_fn, lr, wd, *data)
        train_loss_sum += train_loss[-1]
        valid_loss_sum += valid_loss[-1]
        if i == 0:
            semilogy(range(1, num_epoch+1), train_loss, 'epochs', 'rmse',
                    range(1,num_epoch+1), valid_loss, ['train','valid'])
        print('fold: %d, train_rmse: %.2f, valid_rmse: %.2f'%(i, train_loss[-1], valid_loss[-1]))
    return train_loss_sum/k, valid_loss_sum/k    
```

```python
k_fold(k = 5, train_x = tensor_train_data, train_y = tensor_price, num_epoch = 200, lr = 2.5, wd = 0, batch_size = 64)
```

```python
def fit_and_pred(tensor_train_data_c, tensor_price_c, tensor_test_data_c, test_data_c, num_epoch_c, batch_size_c, lr_c, wd_c):
    model_fn = get_net(tensor_train_data_c.shape[1])
    loss_fn = nn.MSELoss()    
    train_loss, _= train(num_epoch_c, batch_size_c, model_fn, loss_fn, lr_c, wd_c, tensor_train_data_c, tensor_price_c)
    semilogy(range(1, num_epoch_c+1), train_loss, 'epochs', 'rmse')
    print('epoch: %d, train_loss: %.2f'%(num_epoch_c, train_loss[-1]))
    y_pred = model_fn(tensor_test_data_c.float()).detach().squeeze().numpy()
    test_data_c['SalePrice'] = pd.Series(y_pred)
    submission = pd.concat([test_data_c['Id'],test_data_c['SalePrice']], axis = 1)
    submission.to_csv('./Result/submission.csv', index = False)
```
